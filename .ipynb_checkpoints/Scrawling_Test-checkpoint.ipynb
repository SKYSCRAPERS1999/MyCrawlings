```python
import re, jieba
def get_clean_text(data):
    texts = []
    regex_at = re.compile('@([^ |<|:|\.|\*|\|\(|\)|\\|\/|<|>|\[|\]]+)')
    regex_chi = re.compile('[^\u4e00-\u9fa5|，|!|。]+|\?{1}|:{1}')
    for cnt, raw in enumerate(data.text):
        del_name = re.findall(regex_at, raw)
        for name in del_name:
            raw = re.sub(name, '，', raw)
        raw = re.sub(regex_chi, '，', raw)
        raw = re.sub(' +', ' ', raw)
        texts.append(raw)
        
        if cnt % 100 == 0:
            print ('cnt = {}'.format(cnt))
            
    if len(texts) != len(data):
        print('not the same length')
        
    return pd.Series(texts)
```

```python
import re, jieba, thulac

def get_words(data, regex, output=True):
    words = []
    t = thulac.thulac()
    for cnt, raw in enumerate(data.text):
        result = t.cut(raw)
        raw = [x[0] for x in result if (len(x[0]) > 1 and x[1][0] == 'n' and re.search(regex, x[0]) == None)]
        words.append(raw)
        
        if cnt % 100 == 0 and output == True:
            print ('cnt = {}'.format(cnt))
            
    return pd.Series(words)
```

```python
import re
def read_stop_words(): 
    stop_words = []
    stop_letters = []
    with open('stop_words.txt') as sp:
        for x in sp:
            stop_words.append(x[:-1])
    with open('stop_letters.txt') as sp:
        for x in sp:
            stop_letters.append(x[:-1])
    stop_words = set(stop_words)
    re_stop_letter = re.compile('|'.join(stop_letters))
    return (stop_words, re_stop_letter)
stop_words, re_stop_letter = read_stop_words()
```

```python
test_data = pd.read_csv('weibo_test_data.csv')
test_words = test_data['words']
```

```python
def get_word_count(lst):
    count = {}
    for x in eval(lst):
        if x not in stop_words:
            if x in count:
                count[x] += 1
            else:
                count[x] = 1
    return count
```

```python
test_words_dict = [get_word_count(lst) for lst in test_words]
```

```python
import math
def tf(word, count):
    return count[word] / sum(count.values())

def n_containing(word, count_list):
    return sum(1 for count in count_list if word in count)
    
def idf(word, count_list):
    return math.log(len(count_list) / (1 + n_containing(word, count_list)))

def tfidf(word, count, count_list):
    return tf(word, count) * idf(word, count_list)
```

```python
countlist = sorted_idf_vocab
for i, count in enumerate(test_words_dict):
    print("Top words in document {}".format(i + 1))
    scores = {word: tfidf(word, count, countlist) for word in count}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:3]:
        print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
```
